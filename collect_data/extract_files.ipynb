# 1.2: Extract coordinates from pdfs 


```python
# Get data from MongoDB database 
from pymongo import MongoClient

# Replace with your MongoDB connection string
CONNECTION_STRING = "mongodb://127.0.0.1:27018"  # For local MongoDB

# Connect to the database and collection
client = MongoClient(CONNECTION_STRING)

db = client["archaeology_metadata"] # Define database
collection = db["collection"] # Define collection
```

## PAN datasets: extract metadata xmls / zips 

```python
# Find all PAN datasets 
query = {
    "ore:describes.author.citation:authorName": "Portable Antiquities of the Netherlands"}

pan = collection.find(query)

# Grab sample for development
pan = pan.limit(250)
```

```python
# Get all DOIS 
from scripts.json_to_df import get_doi

dois = []
for s in pan: 
    d = get_doi(s)
    dois.append(d)
```

```python
print(len(dois))
```

```python
from scripts.download_ds_files import download_files

selected_files = {"original-metadata.zip/dataset.xml", "easy-migration/dataset.xml"} 

# Get "dataset.xml" or "easy-migration.zip" or "original-metadata.zip"
for doi in dois: 
    download_files(doi, selected_files)
```

## NO-PAN datasets: extract pdf reports 


```python
# Find all non-PAN datasets 
query = {
    "$and": [
        {"ore:describes.dansTemporalSpatial:dansSpatialPoint": {"$exists": False}},
        {"ore:describes.dansTemporalSpatial:dansSpatialBox": {"$exists": False}},
        # condition that the author field is not PAN: 
        {"ore:describes.author.citation:authorName": {"$ne": "Portable Antiquities of the Netherlands"}}
    ]
}
```

```python
nopan = collection.find(query)
```

```python
nopan = nopan.limit(100)
```

```python
from scripts.json_to_df import get_doi

dois = []
for s in nopan: 
    doi = get_doi(s)
    dois.append(doi)
```
