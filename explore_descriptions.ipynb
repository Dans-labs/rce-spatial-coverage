## Named Entity Recognition

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline

# Load the tokenizer and model for RobBERT NER
model_name = "pdelobelle/robbert-v2-dutch-ner"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

# Create a NER pipeline
nlp = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")

# Input text
text = "Ik woon in Amsterdam en werk in Rotterdam."

# Perform Named Entity Recognition
entities = nlp(text)

# Output the results
for entity in entities:
    print(f"Entity: {entity['word']}, Type: {entity['entity_group']}, Score: {entity['score']:.2f}")
```
