# Conversion of Rijksdriehoek to WGS84 coordinates 
Or the other way around 

```python
# Load data
import pandas as pd
df = pd.read_csv('../data/eval_spatialCoverageText.csv')
```

```python
df
```

```python
# Extract coordinates 
all_coords = df.dansSpatialPoints.tolist()
all_coords
```

```python
test = all_coords[:5]
```

```python
from scripts.convert_rd_to_wpg84 import query_api


# Clean responses, turn into dictionary
responses = []
for pair in test:
    coord1, coord2 = pair.split(',')
    coord1 = coord1.strip()
    coord2 = coord2.strip()
    
    response = query_api(coord1, coord2)

    responses.append(response)
```

```python
responses
```

```python
# Convert dansSpatialPoints to WPG84
dansWPG85 = []
for item in responses:
    dansWPG85.append(item['WGS84'])
dansWPG85
```
